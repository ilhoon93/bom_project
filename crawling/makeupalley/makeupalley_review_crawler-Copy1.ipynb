```python
# import module
import numpy as np
import pandas as pd
import re # 스트링 전처리하는 모듈 

# import functions
from bs4 import BeautifulSoup as bs # html에서 어떤 태그로 접근할지 
from requests import get
from time import sleep # sleep : 1 이면 1초 동안 쉬는 것. 너무 빠르게 크롤링하면 봇으로 판단해서 블락당할수 있는데 sleep으로 방지할 수 있음.
```

```python
# import item_id
item_id = pd.read_csv('lipstick.csv') # 바깥쪽 크롤링해서 각각 리뷰들의 url id를 알아내고 그걸 돌려서 리뷰 크롤링 (총 2번 정도해야한다.)
```

```python
item_id.head()
```

```python
# decieve the agent
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} #user agent를 내 pc로 설정 

## 사람인척 속이는 것, 서버에서 막히지 않도록
```

```python
# assign variable
# np data type
dt_i = np.dtype(np.int32) # unsigned integer 4-byte
dt_f = np.dtype(np.float) # unsigned float 2-byte
dt_o = np.dtype(object) # python object

# total 
package_quality_total = np.array([],dtype = dt_f)
price_total = np.array([],dtype = dt_i)

# each
review_rating = np.array([],dtype = dt_i)
review_text = np.array([],dtype = dt_o)
user_id = np.array([],dtype = dt_o)
super_user = np.array([],dtype = dt_o)
time = np.array([],dtype = dt_o)
age = np.array([],dtype = dt_o)
skin_type = np.array([],dtype = dt_o) 
skin_tone = np.array([],dtype = dt_o)
skin_undertone = np.array([],dtype = dt_o)
hair_type = np.array([],dtype = dt_o)
hair_texture = np.array([],dtype = dt_o)
hair_color = np.array([],dtype = dt_o)
eye_color = np.array([],dtype = dt_o)
helpfulness = np.array([],dtype = dt_i)
votes = np.array([],dtype = dt_i)
```

```python
#### crawl
for itemid in item_id.loc[20:,].Item_Id:
    pg = 1
    print('-------------------'+str(itemid)+'----------------------')
    while(1):
        print('page:'+str(pg))
        # initialize
        pqt = []
        pt = []
        rr = []
        rt = []
        ui = []
        su = [] # super : 1 others : 0
        ti = []
        ag = []
        sk_type = []
        sk_tone = []
        sk_undertone = []
        ha_type = []
        ha_texture = []
        ha_color = []
        e_color = []
        helpful = []
        vote = []
        
        # url
        url = 'https://www.makeupalley.com/product/showreview.asp/ItemID='+str(itemid)+'/page='+str(pg)
        html = get(url,headers = headers).text
        soup = bs(html,'html.parser')
        
        # 탈출 조건
        if soup.select('#main > div.col-sm-7.col-md-8.limit-left > div > h1') == []:
            break
        
        #-------------------------------------- html parsing --------------------------------------
        
        
        # review_rating
        for attr in soup.select('div.comment-heading > div.general > div.lipies'):
            for j,child in enumerate(attr.children):
                if j==1:
                    rr.append(int(child['class'][0][2]))
        
        r_num = len(rr) # 리뷰 개수..
        rr = np.array(rr,dtype=dt_i)
        
        
        # total_package_quality, # total_price
        for i,cont in enumerate(soup.select('div > div.col-md-12.col-sm-6.col-xs-6.y-divider ')[0].descendants):
            if soup.select('div > div.col-md-12.col-sm-6.col-xs-6.y-divider > p.pack') == []: # 이상한태그 있으면...
                if i == 11: #pqt
                    pqt.append(float(cont.split()[2]))
                elif i == 13: #price
                    pt.append(int(len(cont.split()[1])))
                else:
                    pass
            else: # 이상한태그 없으면...
                if i == 10: #pqt
                    pqt.append(float(cont.split()[2]))
                elif i == 12: #price
                    pt.append(int(len(cont.split()[1])))
                else:
                    pass
            
            
        
        pqt = np.array(pqt*r_num,dtype = dt_f)
        pt = np.array(pt*r_num,dtype = dt_i)
        
        
        # review_text
        for i in range(r_num):
            [rt.append(re.sub('\t','',review.text)) for i,review in enumerate(soup.select('div.comment-content')[i].descendants) if i==4]
        
        rt = np.array(rt,dtype = dt_o)
        
        # user_id, super_user
        for attr in soup.select('div.comment-heading > div.general > div.user-name'):
            ui.append(re.sub('\W','',attr.text))
            [su.append(re.sub(' ','',child)) for i,child in enumerate(attr.children) if i==2]
    
        ui = np.array(ui, dtype = dt_o)
        su = np.array(su, dtype = dt_o)
        
        # time
        ti = [t.text for t in soup.select('div.comment-heading > div.general > div.date > p')]
        ti = np.array(ti, dtype = dt_o)
        
        # age,sk_type,sk_tone,sk_undertone,ha_type,ha_texture,ha_color,e_color
        for j in range(r_num):
            for k,info in enumerate(soup.select('div.comment-heading > div.important')[j].descendants):
                if k == 1:
                    ag.append(info.text.split()[1])
                elif k == 9:
                    sk_type.append(re.sub(':','',info).split(',')[0])
                    sk_tone.append(re.sub(':','',info).split(',')[1])
                    sk_undertone.append(re.sub(':','',info).split(',')[2])
                elif k == 14:
                    ha_type.append(re.sub(':','',info).split(',')[0])
                    ha_texture.append(re.sub(':','',info).split(',')[1])
                    ha_color.append(re.sub(':','',info).split(',')[2])
                elif k == 19:
                    e_color.append(re.sub(':','',info).split(',')[0])
                else:
                    pass
                
        ag = np.array(ag,dtype = dt_o)
        sk_type = np.array(sk_type,dtype = dt_o)
        sk_tone = np.array(sk_tone,dtype = dt_o)
        sk_undertone = np.array(sk_undertone,dtype = dt_o)
        ha_type = np.array(ha_type,dtype = dt_o)
        ha_texture = np.array(ha_texture,dtype = dt_o)
        ha_color = np.array(ha_color,dtype = dt_o)
        e_color = np.array(e_color,dtype = dt_o)
        
        #helpfulness
        for thumb in soup.select('div.thumbs > p'):
            if thumb.text == '':
                helpful.append(0)
                vote.append(0)
            else:
                helpful.append(int(re.sub('[^0-9]','',thumb.text)[0]))
                vote.append(int(re.sub('[^0-9]','',thumb.text)[1]))
        
        helpful = np.array(helpful,dtype = dt_i)
        vote = np.array(vote,dtype = dt_i)
        
        #stack
        package_quality_total = np.hstack((package_quality_total,pqt))
        price_total = np.hstack((price_total,pt))
        review_rating = np.hstack((review_rating,rr))
        review_text = np.hstack((review_text,rt))
        user_id = np.hstack((user_id,ui))
        super_user = np.hstack((super_user,su))
        time = np.hstack((time,ti))
        age = np.hstack((age,ag))
        skin_type = np.hstack((skin_type,sk_type))
        skin_tone = np.hstack((skin_tone,sk_tone))
        skin_undertone = np.hstack((skin_undertone,sk_undertone))
        hair_type = np.hstack((hair_type,ha_type))
        hair_texture = np.hstack((hair_texture,ha_texture))
        hair_color = np.hstack((hair_color,ha_color))
        eye_color = np.hstack((eye_color,e_color))
        helpfulness = np.hstack((helpfulness,helpful))
        votes = np.hstack((votes,vote))
        
        
        pg += 1
    
```
