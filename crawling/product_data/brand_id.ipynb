```python
import pandas as pd
from pandas import Series, DataFrame
import re
import numpy as np
from requests import get
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome('C:/Users/Ilhoon/desktop/dev_util/chromedriver_win32/chromedriver')
driver.get('http://glowpick.com/brand/list')
```

```python
brand_name=[]
brand_id=[]
```

```python
for i in range(2,17):
    driver.find_element_by_xpath('//*[@id="gp-brand-list"]/div/section[1]/form/fieldset[2]/ul/li['+str(i)+']').click()
    time.sleep(1)
    html = driver.page_source
    soup=bs(html,'lxml')
    for k in range(len(soup.select('div.comp-card-simple'))):
        brand_name.append(soup.select('div.comp-card-simple > div.simple-wrap')[k].select('p.simple-name > span.txt')[1].text)
        brand_id.append(re.sub("[^0-9]","",soup.select('div.comp-card-simple > a')[k].get('href')))
```

```python
data={'brand_name':brand_name,
     'brand_id':brand_id}
brand_data=DataFrame(data)
brand_data=DataFrame(data, columns=['brand_id','brand_name'])
```
