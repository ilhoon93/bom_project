```python
# 데이터 저장할 리스트
score_dic={"짱짱!":5,"굿굿!":4,"쏘쏘!":3,"별로!":2,"최악!":1} # 별점 딕셔너리
user_name=[] # 사용자 이름
age=[] # 나이
skin_type=[] # 피부타입
score=[] # 별점
created_date=[]
review_text=[] # 리뷰 텍스트
item_name=[]
```

```python
# 하나의 리뷰가 2장 이상 넘어가는 경우
# 날짜의 형태가 다름 .으로 구분, -으로 구분

#  데이터 전처리 함수
def manipulate(data):
    review=data # 리뷰 텍스트만 담을 변수 
    review.replace("(g","")
    review.replace("(c","")
    
    item_name.append("디스어포어 뷰티 밤")
    # 사용자 이름
    try:
        if "(" in data[:20]:
            a= data.find("(")
            user_name.append(data[:a]) # 사용자 이름 user_name에 저장 
            review.replace(data[:a],"")
            review=re.sub(data[:a],"",review)
        else: # 맨 처음 여는 괄호를 제대로 인식하지 못한 경우
            a=data.find("/")
            user_name.append((data[:a-3])) 
            review=re.sub(data[:a-3],"",review)
    except:
        user_name.append("")
        
    # 나이
    try:
        c= data.find("/")
        age.append(re.sub("[^0-9]","",data[a+1:c])[:2]) # 나이 age에 저장
        review=re.sub(re.sub("[^0-9]","",data[a+1:c])[:2],"",review)
    except:
        age.append("")
    
    
    # 피부타입
    try:
        b= data.find(")")
        skin_type.append(re.sub("\s","",data[c+1:b])) # 피부 타입 skin_type에 저장 
        review=re.sub(data[c+1:b],"",review)
        review=re.sub("\)","",review)
        review=re.sub("\(","",review)
        review=re.sub("/","",review)
        q=data.find("세")
        review=re.sub(data[q],"",review) # 나이 x세 "세" 삭제
    except:
        skin_type.append("")

    # 평점
    wrong_score=["쏘씨","쏘시","쏘쐬","쏘쏘,","쏘쏘.","쏘쏘:","쏘씬","또소","4씨","산산!"]
    for text in wrong_score:
        data = data.replace(text,"쏘쏘!")
    wrong_score1=["별뢰","@뢰","@로!","벌로!","별로,","@로","벌뢰","벨로","빌로"]
    for text in wrong_score1:
        data=data.replace(text,"별로!")
    wrong_score2=["짱짱","방쌍!","빵빵!","빵짱",'방쌩',"방쌍","방쨩","방짱","방쨍","빵빵","방생","빵생"]
    for text in wrong_score2:
        data=data.replace(text,"짱짱!")
    wrong_score3=["굿굿","굿귄","굿귓","a귄"]
    for text in wrong_score3:
        data=data.replace(text,"굿굿!")
    wrong_score4=['최액',"최악",'화악!']
    for text in wrong_score4:
        data=data.replace(text,"최악!")
   




    if "짱짱!" in data:
        d=data.find("짱짱!")
        score.append(score_dic[data[d:d+3]])
        review=re.sub(data[d:d+3],"",review)
    elif "굿굿!" in data:
        d=data.find("굿굿!")
        score.append(score_dic[data[d:d+3]])
        review=re.sub(data[d:d+3],"",review)
    elif "쏘쏘!" in data:
        d=data.find("쏘쏘!")
        score.append(score_dic[data[d:d+3]])
        review=re.sub(data[d:d+3],"",review)
    elif "별로!" in data:
        d=data.find("별로!")
        score.append(score_dic[data[d:d+3]])
        review=re.sub(data[d:d+3],"",review)
    elif "최악!" in data:
        d=data.find("최악!")
        score.append(score_dic[data[d:d+3]])
        review=re.sub(data[d:d+3],"",review)
    else:
        score.append("") # 없으면 공백


# 날짜
    year=int(str(datetime.today())[:4]) # 오늘 년
    month=int(str(datetime.today())[5:7]) # 오늘 월
    day=int(str(datetime.today())[8:10]) # 오늘 일 
    review=re.sub("시간 전","",review)
    review=re.sub("일 전","",review)
    review=re.sub("개월 전","",review)


    p=re.compile('[0-9]{4}[.][0-9]+[.][0-9]+') # 날짜 텍스트를 정확히 인식했다면 
    p1=re.compile('[0-9]{3}[.][0-9]{2}[.][0-9]{2}') # 하나 잘못 인식
    m=p.search(data)
    m1=p1.search(data)

    if m!= None:
        created_date.append(m.group())
        review=re.sub('[0-9]+[.][0-9]+[.][0-9]+',"",review)
    elif m1!= None:
        tmpp=m1.group()
        m_m="2"+tmpp
        created_date.append(m_m)
        review=re.sub('[0-9]+[.][0-9]+[.][0-9]+',"",review)
        review=re.sub('[0-9]+[.][0-9]+[.]\s',"",review)

    elif "시간 전" in data:
        created_date.append(date(year, month, day)) # 몇 시간 전 정보는 모두 오늘 날짜로 처리 
        f=data.find("시간 전")
        n_time=re.sub("[^0-9]","",data[f-3:f])
        review=re.sub(n_time,"",review)
    elif "일 전" in data:
        e= data.find("일 전")
        d_day=int(re.sub("[^0-9]","",data[e-3:e])) # 몇일 전 인지
        review=re.sub(str(d_day),"",review)
        if (day-d_day)>=1:
            n_date=date(year,month,day-d_day)
            created_date.append(n_date)
        else:
            lastday=calendar.monthrange(year,month-1)[1]
            temp=lastday-d_day
            n_date=date(year,month-1,temp+day)
            created_date.append(n_date)
    elif "개월 전" in data:
        z=data.find("개월 전")
        n_month=int(re.sub("[^0-9]","",data[z-3:z])) # 몇 달전 인지
        review=re.sub(str(n_month),"",review) # 몇 달 전 숫자 빼주기
        if (month-n_month)>=1:
            n_date=date(year,month-n_month,day)
            created_date.append(n_date)
        else:
            temp=12-n_month
            n_date=date(year,temp+month,day)
            created_date.append(n_date)
    else:
        created_date.append("")


    # 리뷰 텍스트 
    try:
        wrong=[]
        wrong_char="よソơởσこ②つじ人。のしのしっ℃ợ€瓷ж週わ黜四甜器售ⓤ靉づ阖" # 잘못 들어온 데이터 들을 wrong에 저장 
        wrong.extend(wrong_char)
        wrong.extend(wrong_score)
        wrong.extend(wrong_score1)
        wrong.extend(wrong_score2)
        wrong.extend(wrong_score3)
        wrong.extend(wrong_score4)
        for char in wrong:
            review = review.replace(char,"")

        review=review.replace("π{2,}","ㅠ")# 파이를 ㅠㅠ로 변환
        review=review.replace("T{2,}","ㅜ")
        review=review.replace("L{2,}","ㄴ")
        review=review.replace("t\s{1}[0-9]","") 

        if "i"or"t"or"g"or"c"or"S"or"0" in review[-5:]:
            review=re.sub("i","",review)
            review=re.sub("t","",review)
            review=re.sub("g","",review)
            review=re.sub("c","",review)
            review=re.sub("S","",review)
            review=re.sub("0","",review)
        review=review.replace("1b1","")
        review=review.replace("lb1","")
        review=review.replace("0by,","")
        review=re.sub("\s+[0-9]\s+","",review)

        if review[0]=="4":
            review=re.sub("4","",review)
            
        if "d"or"d."or"d'"or"H"or"o" in review[:3]:
            review=re.sub("d","",review)
            review=re.sub("d'","",review)
            review=re.sub("d.","",review)
            review=re.sub("H","",review)
            review=re.sub("o","",review)

        review_text.append(review.strip())
    except:
        review_text.append("")
```
