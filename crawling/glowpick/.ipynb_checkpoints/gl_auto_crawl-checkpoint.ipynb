```python
# bb,cc,파운데이션 상위 100개만 크롤링하는 코드 - 프린트

def t_gl_100_crawl():
    dt=datetime.now() # 현재 날짜 dt에 저장
    
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    
 #   urls = {"http://glowpick.com/beauty/ranking?id=42":"bb크림",     # bb크림 
 #          "http://glowpick.com/beauty/ranking?id=136":"cc크림",      # cc크림
 #         "http://glowpick.com/beauty/ranking?id=43":"파운데이션"}  # 파운데이션
    
    for product in gl_100_id:
        url='http://glowpick.com/product/'+str(product)+''
        html=get(url,headers=headers).text
        soup = bs(html, 'html.parser')
        
        for k in range(len(soup.select('section.section-list-item > ul.review-list-wrap > li'))): # 임시로 제품 길이만큼만 최대 20개
            
            print(product) # 제품 아이디
            print(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-top > li.section-list-info.side-right > div > section.section-list-item.product > h1.pdt-name')[0].text) # 제품명 보류 

            # 판매처
            try:
                print(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li.sale > div > ul > li')[0].text)
            except:
                print("")
            # 제품 속성
            product_attribute=[]
            for j in range(len(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li > ul > li'))):
                print(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li > ul > li')[j].text)
            print(",".join(product_attribute)) 

            # 제품 설명 (콤마로 구분) 
            print(re.sub("\r+",",",re.sub("\n+","",soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li.desc > p')[0].text.strip())))

            # 피부타입
            print(re.sub("·","",soup.select('section.section-list-item > ul.review-list-wrap > li > div > div > div > p > span.info > span.txt')[k].text[4:]).strip())
            # 나이
            print(soup.select('section.section-list-item > ul.review-list-wrap > li > div > div > div > p > span.info > span.txt')[k].text[:3])
            # 유저 이름
            print(soup.select('section.section-list-item > ul.review-list-wrap > li > div > div > div > p > span.name')[k].text)
            # 리뷰 텍스트
            print(soup.select('section.section-list-item > ul.review-list-wrap > li > div > p')[k].text)
            # 날짜 정보
            print(soup.select('section.section-list-item > ul.review-list-wrap > li > span.date')[k].text)
        print("\n\n")
        
        
    
```

```python
# 브랜드 id로 각각 url 접근
brand_name=[] # 브랜드명
product_name=[] # 브랜드이름
product_id=[] # 제품 ID
product_volume=[] # 제품 용량
product_price=[] # 제품 가격
product_score=[] # 제품 평점
product_score_joiner=[] # 제품 평점 참여자 수
```

```python
## 순서
# 0. 제품별 리뷰 개수를 포함해 리뷰 재 크롤링 
# 1. product_data에 제품 category join
# 2. product_data에 제품 category가 bb,cc,파운데이션인 것들만 뽑아냄
# 3. id 별로 url 접근 
# 4. 첫페이지에 있는 것들만 실시간으로 저장( 저장된 리뷰 개수보다 방금 뽑아낸 리뷰 개수가 클 때만 큰 만큼 크롤링 반복)


def gl_crawl():
    dt=datetime.datetime.now() # 현재 날짜 dt에 저장
    
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    
    urls = {"http://glowpick.com/beauty/ranking?id=42":"bb크림",     # bb크림 
           "http://glowpick.com/beauty/ranking?id=136":"cc크림",      # cc크림
           "http://glowpick.com/beauty/ranking?id=43":"파운데이션"}  # 파운데이션
    
    for url in urls:
    
        driver.get(url)
        for i in range(30):
            screenWidth, screenHeight = pyautogui.size()
            currentMouseX, currentMouseY = pyautogui.position()
            pyautogui.moveTo(1910, 1008.5)
            pyautogui.click()

        
        html = get(url, headers=headers).text
        soup = bs(html, 'html.parser')
        
        for k in range(len(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info'))): # 브랜드별 아이템정보 담고 있는 개체
            brand_name.append(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info')[k].select('div.side-prod > h5.brand')[0].text) # 브랜드 명
            product_name.append(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info')[k].select('div.side-prod > h6.name')[0].text) # 제품 명
            product_id.append(re.sub("[^0-9]","",soup.select('div.comp-card-prod')[k].get('data-url'))) # 제품 id 
            product_volume.append(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info')[k].select('div.side-prod > p > span.volume')[0].text) # 용량
            product_price.append(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info')[k].select('div.side-prod > p > span.price')[0].text) # 가격
            product_score.append(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info')[k].select('div.side-score > div > span.score')[0].text) # 평점
            product_score_joiner.append(soup.select('div.comp-card-prod > div.prod-wrap > div.side-info')[k].select('div.side-score > div > span.joiner')[0].text[1:-1]) # 평점 참여자 수
```

```python
# 글로우픽 제품 리뷰

gl_product_id=[] # 제품 id
gl_product_attribute=[] # 제품 속성
gl_sales=[] # 판매처
gl_product_desc=[] # 제품 설명
gl_user_name=[] # 유저 이름
gl_age=[] # 나이
gl_skin_type=[] # 피부 타입
gl_date=[] # 날짜
gl_review=[] # 리뷰 텍스트
```

```python
# 제품 별 리뷰 데이터 크롤링
j=1
for product in df_glow_id["product_id"]:
    
    driver.get('http://glowpick.com/product/'+str(product)+'')
    time.sleep(1)
    print(j)
    j+=1
    for i in range(50):
        screenWidth, screenHeight = pyautogui.size()
        currentMouseX, currentMouseY = pyautogui.position()
        pyautogui.moveTo(1910, 1008)
        if i==30:
            time.sleep(1)
        pyautogui.click()

    html = driver.page_source
    soup = bs(html, 'html.parser')

    for k in range(len(soup.select('section.section-list-item > ul.review-list-wrap > li'))):
        # 제품 id 
        gl_product_id.append(product)
        # 판매처
        try:
            gl_sales.append(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li.sale > div > ul > li')[0].text)
        except:
            gl_sales.append("")
        # 제품 속성
        product_attribute=[]
        for j in range(len(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li > ul > li'))):
            product_attribute.append(soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li > ul > li')[j].text)
        gl_product_attribute.append(",".join(product_attribute)) 

        # 제품 설명 (콤마로 구분) 
        gl_product_desc.append(re.sub("\r+",",",re.sub("\n+","",soup.select('#gp-product-detail > div > ul.section-list-wrap.side-bottom > li.section-list-pdt.side-left > section > div > ul > li.desc > p')[0].text.strip())))
        
        # 피부타입
        gl_skin_type.append(re.sub("·","",soup.select('section.section-list-item > ul.review-list-wrap > li > div > div > div > p > span.info > span.txt')[k].text[4:]).strip())
        # 나이
        gl_age.append(soup.select('section.section-list-item > ul.review-list-wrap > li > div > div > div > p > span.info > span.txt')[k].text[:3])
        # 유저 이름
        gl_user_name.append(soup.select('section.section-list-item > ul.review-list-wrap > li > div > div > div > p > span.name')[k].text)
        # 리뷰 텍스트
        gl_review.append(soup.select('section.section-list-item > ul.review-list-wrap > li > div > p')[k].text)
        # 날짜 정보
        gl_date.append(soup.select('section.section-list-item > ul.review-list-wrap > li > span.date')[k].text)
```

```python
data={'product_id':gl_product_id,
      'product_attribute':gl_product_attribute,
     'sales':gl_sales,
     'product_desc':gl_product_desc,
     'user_name':gl_user_name,
     'age':gl_age,
     'skin_type':gl_skin_type,
     'date':gl_date,
     'review':gl_review}
gl_review=DataFrame(data)
gl_review=DataFrame(data, columns=['product_id','product_attribute','sales','product_desc','user_name','age','skin_type','date','review'])
gl_review
```
