```python
import pandas as pd
from pandas import Series, DataFrame
import re
import numpy as np

from requests import get
from bs4 import BeautifulSoup as bs
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
```

```python
meme_bb_id = pd.read_csv('meme_bb_id.csv')
meme_bb_id
```

```python
def starnum(star): # 별의 텍스트를 뽑아내는 함수
    first=0
    second=0
    for i in range(len(star)):
        if star[i]=="\"":
            if first==0:
                first=i
            else: 
                second=i
    return star[first+1:second]
```

```python
score_dic ={"":0,"half":0.5,"one":1,"oneHalf":1.5,"two":2,"twoHalf":2.5,"three":3,"threeHalf":3.5,"four":4,"fourHalf":4.5,"five":5} 
    # 별점이 숫자가 아닌 텍스트로 되어있음. 텍스트에 대응되는 별점 딕셔너리
```

```python
# 기본 화면에서 review데이터 들어오지 않음
# F12 네트워크 찾아서 실시간 업로드 되는 파일 검색
# 리뷰 관련 파일 찾아서 URL 접속
# http://www.memebox.com/ajax/reviews?productId=511587&page=2&contentViewPickOption=2&hasBest=true
# page=x 식으로 넣어서 페이지 찾음
item_id=[]
review_text=[] # 리뷰 텍스트
created_date=[] # 리뷰 작성일
star_score=[] # 별점 
like_thumb_count=[] # 리뷰 좋아요 개수
skin_type=[] # 피부타입
repurchase=[] # 재구매여부
```

```python
# test
for i in meme_bb_id['item_id']:
    turl='http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page=1&contentViewPickOption=2&hasBest=true'
    thtml = get(turl,headers=headers).text
    tsoup=bs(thtml,'html.parser')
    if 50 < len(str(tsoup.select('div > a.arrow.last'))):
        ReviewPageNum=tsoup.select('div > a.arrow.last')[0].get('onclick')[15:17]
        ReviewPageNum=re.sub("'","",ReviewPageNum)
    else:
        ReviewPageNum=1
        
        # 리뷰 페이지 총 몇 개 인지
        # 텍스트 길이로 리뷰 페이지 1 인지 그 이상인지 구분
        # "9'" 잘못 들어온 홑따옴표 제거
        # 리뷰 페이지 수를 얻기 위한 임시 url,html, soup
        
    for j in range(1,int(ReviewPageNum)+1): # item별 리뷰 페이지의 수 만큼 반복
        url = 'http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page='+str(j)+'&contentViewPickOption=2&hasBest=true'
        html = get(url,headers=headers).text
        soup=bs(html,'html.parser')
        
        for k in range(len(soup.select('div > p.reviews-written-text-deal-detail'))): # 리뷰 페이지별 리뷰 수 만큼 반복
            print("item_id " +soup.select('div.view')[k].get('data-product-id')) # item_id
            print("review_text "+re.sub('\t','',soup.select('div > p.reviews-written-text-deal-detail')[k].text).strip()) # 리뷰 텍스트
            for i,child in enumerate(soup.select('div.info > span')[k]):
                if len(child)==4:
                    skin_type.append(re.sub("/","",child).strip())
                else:
                    skin_type.append("")
                         
            print("created_date "+ soup.select('div.info > span.date')[k].text) # 리뷰 작성일 
            print("star_score %d" % (score_dic[starnum(str(soup.select('span.star > em')[k]))])) # 리뷰 평점
            print("like_thumb_count "+re.sub("\n| ","",soup.select('span.like-thumb-text')[k].text)) # 리뷰 좋아요 개수 
            
```

```python
# run crawling
for i in meme_bb_id['item_id']:
    turl='http://www.memebox.com/ajax/reviews?productId='+str(i)+'&contentViewPickOption=2&hasBest=true'
    thtml = get(turl,headers=headers).text
    tsoup=bs(thtml,'html.parser')
    if 50 < len(str(tsoup.select('div > a.arrow.last'))):
        ReviewPageNum=tsoup.select('div > a.arrow.last')[0].get('onclick')[15:17]
        ReviewPageNum=re.sub("'","",ReviewPageNum)
    else:
        ReviewPageNum=1
        
        # 리뷰 페이지 총 몇 개 인지
        # 텍스트 길이로 리뷰 페이지 1 인지 그 이상인지 구분
        # "9'" 잘못 들어온 홑따옴표 제거
        # 리뷰 페이지 수를 얻기 위한 임시 url,html, soup
        
    for j in range(1,int(ReviewPageNum)+1): # item별 리뷰 페이지의 수 만큼 반복
        url = 'http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page='+str(j)+'&contentViewPickOption=2&hasBest=true'
        html = get(url,headers=headers).text
        soup=bs(html,'html.parser')
        
        for k in range(len(soup.select('div > p.reviews-written-text-deal-detail'))): # 리뷰 페이지별 리뷰 수 만큼 반복
            item_id.append(soup.select('div.view')[k].get('data-product-id')) # item_id
            review_text.append(re.sub('\t','',soup.select('div > p.reviews-written-text-deal-detail')[k].text).strip()) # 리뷰 텍스트
            
            if len(soup.select('div.info')[k].select('span'))==2:
                skin_type.append(re.sub("/","",soup.select('div.info')[k].select('span')[0].text).strip()) #  피부타입 
                created_date.append(soup.select('div.info')[k].select('span')[1].text)
            else:
                skin_type.append("") # 피부타입 작성 안했으면 공백
                created_date.append(soup.select('div.info')[k].select('span')[0].text) # 리뷰 작성일 
     
            star_score.append(score_dic[starnum(str(soup.select('span.star > em')[k]))]) # 리뷰 평점
            like_thumb_count.append(re.sub("\n| ","",soup.select('span.like-thumb-text')[k].text)) # 리뷰 좋아요 개수 
                # 정규 표현식 "\n| "으로 \n 혹은 공백을 제거
            if soup.select('div.review-selected-option')[k].select('span'):
                repurchase.append(soup.select('div.review-selected-option')[k].select('span')[0].text) # 재구매 여부
            else:
                repurchase.append("")
```

```python
data ={'item_id':item_id,
      'review_text':review_text,
      'created_date':created_date,
      'star_score':star_score,
      'like_thumb_count':like_thumb_count,
       'skin_type':skin_type,
       'repurchase':repurchase
    }
meme_bb_review=DataFrame(data)
meme_bb_review=DataFrame(data, columns=["item_id",'review_text','created_date','star_score','like_thumb_count','skin_type','repurchase'])
meme_bb_review
```
