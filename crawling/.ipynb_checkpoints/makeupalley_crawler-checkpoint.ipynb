# makeupalley_crawl

### 1. makeupalley 소개
- 크게 product_reviews/ boards/ Favorites 섹션이 있다.
- product_reviews에는 총 2,722,105 reviews of 167,901 products의 리뷰가 있다.
- 그 중 lipstick 카테고리가 __제품 종류__가 제일 많으니까 이걸로 긁음.(29,192 products)

### 2. 크롤링 계획
__url 구조__<br>
(1) `https://www.makeupalley.com/product/searching.asp/CategoryID=`(카테고리 번호)`/dsc=1`(숨김해제)`/page=`(페이지 번호)`/`
<br><br>
(2)`https://www.makeupalley.com/product/showreview.asp/ItemId=`(아이템 번호)`/page=`(페이지 번호)`/`
<br><br>
_(1)을 통해 (2)에 해당하는 아이템 번호를 알아내야됨._

<br>

__1차 크롤링__<br>
lip_url = `https://www.makeupalley.com/product/searching.asp/CategoryID=`2`/dsc=1/page=`i`/`<br>
lip_url을 iteration해서 lipstick 카테고리의 모든 페이지에 등장하는 lipstick 종류의 itemid를 알아내는 것이 목적

### 3. makeupalley 크롤링 issue

- 리뷰 수가 5개 이상인 것만 긁자. (b/c 리뷰가 너무 적으면 의미있는 분석하기 어려움, 29,192 -> 12,281로 우선 중심적인 의미에 집중할 수 있음.)
- 사용자가 많아서 크롤링하는 도중에 사이트가 계속 변화함... $\longrightarrow$ 우선 리뷰 수를 desc로 하자. 그러고 계속 쌓자.

# 1차 Crawl Code

```python
# import module
import numpy as np
import pandas as pd
import re

# import functions
from bs4 import BeautifulSoup as bs
from requests import get
from time import sleep
```

```python
####--------- fuck_alley ------------
# 설명: itemid를 크롤링하기 어렵게 만든 ㅈ같은 makeupalley를 조지는 함수
# category_id : 원하는 제품 종류의 category_id
# NoR : 최소 리뷰의 수 
# sd : DESC or ASC

def fuck_alley(c_id,NoR,sd):
    # decieve the agent
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} #user agent를 내 pc로 설정 
    
    # number of product
    p_num = 990 # 어짜피 앨리는 한번에 최대한 990개의 product만 볼 수 있다.
    
    # np data type
    dt_i = np.dtype(np.int32) # unsigned integer 4-byte
    dt_f = np.dtype(np.float) # unsigned float 2-byte
    dt_o = np.dtype(object) # python object

    # int container
    item_id = np.arange(p_num,dtype=dt_i).reshape(p_num,1) # p_num 길이의 array 만들고 (p_num,1)의 column vecotr로 변환
    review_num = np.arange(p_num,dtype=dt_i).reshape(p_num,1)

    # float container
    rating = np.arange(p_num,dtype=dt_f).reshape(p_num,1)
    buy_again = np.arange(p_num,dtype=dt_f).reshape(p_num,1)

    # object container
    brand = np.arange(p_num,dtype=dt_o).reshape(p_num,1)
    product = np.arange(p_num,dtype=dt_o).reshape(p_num,1)
    category = np.arange(p_num,dtype=dt_o).reshape(p_num,1)
    
    
    #### crawl!!!
    i = 1
    stack_count = 0
    while(1):
        ## page만 바꿀거임
        url = 'https://www.makeupalley.com/product/searching.asp/CategoryId='+str(c_id)+'/NumberOfReviews='+str(NoR)+'/dsc=1/SD='+str(sd)+'/SC=HOWMANY/page='+str(i)
        html = get(url,headers = headers).text
        soup = bs(html,'html.parser')

      
        # soup에 더이상 table이 없으면 끝!
        if soup.select('div > table') == []:
            break

        ## ith page의 table에 있는 값 모두 crawl
        for j, child in enumerate(soup.table.children):
            if j != 0 : # j=0은 table head 값            
                item_id[stack_count] = np.asarray(int(re.sub('[^0-9]','',list(child.children)[0].select('a')[0].attrs['href'].split('/')[3])),dtype = dt_i).reshape(1,1) # itemid
                brand[stack_count] = np.asarray(list(child.children)[1].text,dtype=dt_o).reshape(1,1) # brand
                product[stack_count] = np.asarray(list(child.children)[2].select('a')[1].text,dtype=dt_o).reshape(1,1) # product
                category[stack_count] = np.asarray(list(child.children)[3].text,dtype=dt_o).reshape(1,1) # category
                rating[stack_count] = np.asarray(float(list(child.children)[4].text),dtype=dt_f).reshape(1,1) # rating
                review_num[stack_count] = np.asarray(int(re.sub('[^0-9]','',list(child.children)[5].text)),dtype=dt_i).reshape(1,1) # review_num
                buy_again[stack_count] = np.asarray(float(re.sub('[^0-9]','',list(child.children)[6].text))/100,dtype=dt_f).reshape(1,1) # buy again
                stack_count += 1 # array의 index

        print(stack_count) # number of item!!!
        i += 1
        sleep(1)
    ####
    # 1-d arrays to dataframe
    result = pd.DataFrame(np.hstack((item_id,brand,product,category,rating,review_num,buy_again)))
    result.columns = ['Item_Id','Brand','Product','Category','Rating','Reviews','Buy_Again']
    
    return result
```

```python
def find_asc_final(c_id):
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} #user agent를 내 pc로 설정 
    url = 'https://www.makeupalley.com/product/searching.asp/CategoryId='+str(c_id)+'/NumberOfReviews=5/dsc=1/SD=DESC/SC=HOWMANY/page=66'
    html = get(url,headers = headers).text
    soup = bs(html,'html.parser')
    
    ## 
    for j, child in enumerate(soup.table.children):
        if j == 14 : # j=14은 table 마지막
            review_num = int(re.sub('[^0-9]','',list(child.children)[5].text))

    return review_num
```

```python
#### DESC_crawl
category = 2
NoR = 5 #최소 리뷰
sd = 'DESC'

#한번돌리기...
result_desc = fuck_alley(category,NoR,sd)
asc_final = find_asc_final(category)
```

```python
#### ASC_crawl
sd='ASC'
result_asc = pd.DataFrame(None)

for i in range(5,asc_final+1): # min review = 5 max = asc_final
    asc_temp = fuck_alley(category,i,sd)
    result_asc = pd.concat((result_asc,asc_temp),axis=0)
    result_asc.drop_duplicates(inplace=True)
    print('--------------------------------'+str(i)+'--------------------------------')
```

```python
# 마무리
total_result = pd.concat((result_desc,result_asc),axis=0)
total_result.drop_duplicates(inplace=True)
total_result = total_result.sort_values(by=['Reviews'],axis=0,ascending=False)

# 저장
total_result.to_csv('lipstick.csv',index=False)
```
