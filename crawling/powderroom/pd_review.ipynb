```python
import pandas as pd
import re
import requests
import json
from pandas import DataFrame,Series
from bs4 import BeautifulSoup
from requests import get
from time import sleep
import math
from mod import *
import datetime
# from requests.adapters import HTTPAdapter
# from requests.packages.urllib3.util.ssl_ import create_urllib3_context
```

```python
# decieve the agent // 브라우져 바꿔가면서 크롤링하게끔
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} #user agent를 내 pc로 설정 
```

```python
# load data
pd_id = pd.read_csv('pd_id.csv')
```

```python
pd_id.drop_duplicates(['pid','pname'])
```

```python
c_list={100:"모공",200:"각질",300:"주름",400:"잡티",500:"건조함",600:"피부톤",700:"민감성",800:"트러블",900:"아토피",1000:"피지과다"}
s_type={100:"건성",200:"중성",300:"지성",400:"복합성"}
```

```python
# stack data

# product 
pname = []
pid = []
brandname=[]
category=[]

# user
age=[]
concern_list=[]
skin_type=[]
repurchased=[]

#review
d_date=[]
score=[]
like_c=[]
tags=[]
text=[]
```

```python
# run crawl

for p_id in pd_id['pid']:
    url_p='https://www.powderroom.co.kr/api/products/'+str(p_id)+''
    html_p = get(url_p, headers = headers).text
    products = json.loads(html_p)
    page=math.ceil(int(pd_id.loc[pd_id.pid==int(p_id),'num_of_reviews'].values)/20) # 페이지수 한 페이지당 20개
    schaft = ''
    
    for pn in range(page):
        url='https://www.powderroom.co.kr/api/products/'+str(p_id)+'/reviews?order=new'+ str(schaft)+''
        html = get(url, headers = headers).text
        reviews = json.loads(html)

        # 페이지 당 리뷰 20번 반복, 마지막 페이지 라면 리뷰 개수만큼
        nk=20
        if pn==(page-1):
            nk=int(pd_id.loc[pd_id.pid==int(p_id),'num_of_reviews'].values)%20

        # 페이지당 리뷰 개수만큼 반복
        for i in range(nk):

            ### product
            pid.append(p_id) # 제품 id
            pname.append(str(pd_id.loc[pd_id.pid==int(p_id),'pname'].values)[2:-2]) # 제품 이름
            brandname.append(str(pd_id.loc[pd_id.pid==int(p_id),'brandname'].values)[2:-2]) #브랜드 이름
            category.append(str(pd_id.loc[pd_id.pid==int(p_id),'category'].values)[2:-2]) #카테고리

            ### review 
            date_result=pd.to_datetime(str(reviews[i]['$created'])[:-3],unit='s')
            d_date.append(date_result.strftime('%Y-%m-%d'))
            # print(reviews[i]['$created']) # date 
            score.append(reviews[i]['rate'])# score 
            like_c.append(reviews[i]['count']['like'])# like_c
            tags.append(reviews[i]['tags'])# tags 
            text.append(reviews[i]['text']) # 텍스트 


            ### user        
            try: # json에 age 키 값이 없는 경우가 있다. (사이트에는 있음)
                age.append(reviews[i]['user']['age']) # 나이 
            except:
                age.append(" ")
            try: # 피부 타입
                skin_type.append(s_type[reviews[i]['user']['skin']['type']])# skin_type
            except:
                skin_type.append(" ") 

            tmp_c_list=[]
            try: # 피부고민, 
                for a in reviews[i]['user']['skin']['concerns']:
                    tmp_c_list.append(c_list[a])
                concern_list.append(",".join(tmp_c_list))
            except:
                concern_list.append(" ") 

            repurchased.append(reviews[i]['repurchased'])# repurchase     
        schaft = [rev['_searchAfter'] for rev in reviews][-1]
```

```python
print(len(pid))
print(len(pname))
print(len(brandname))
print(len(age))
print(len(tags))
print(len(repurchased))
print(len(like_c))
print(len(skin_type))
print(len(score))
print(len(d_date))
print(len(text))
print(len(concern_list))
```

```python
data = pd.DataFrame(
    {'brandname': brandname,
    'pid': pid,
     'pname' : pname,
     'age':age,
     'tags':tags,
     'repurchased':repurchased,
     'like_c':like_c,
     'concern_list':concern_list,
     'skin_type':skin_type,
     'score':score,
     'date':d_date,
     'text':text
    })
pd_review=DataFrame(data)
pd_review=DataFrame(data, columns=['brandname',"pid",'pname','age','tags','repurchased','like_c','concern_list','skin_type','score','date','text'])
pd_review
```
