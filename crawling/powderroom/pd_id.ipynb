```python
import pandas as pd
import re
import requests
import json

from pandas import DataFrame,Series
from bs4 import BeautifulSoup as bs
from requests import get
from time import sleep
# from requests.adapters import HTTPAdapter
# from requests.packages.urllib3.util.ssl_ import create_urllib3_context
```

```python
# decieve the agent // 브라우져 바꿔가면서 크롤링하게끔
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'} #user agent를 내 pc로 설정 
```

```python
pid=[] 
pname=[]
brandname=[]
p_score=[]
num_of_reviews=[]
category=[]
```

```python
# 정렬 기준
pd_category={"C2300":'bb크림',
                "C2400":'cc크림',
                'C2500':'파운데이션'}
standards=['best','review','rate','highPrice','lowPrice']
```

```python
for ctr in pd_category:
    for std in standards:
        sleep(1)
        url='https://www.powderroom.co.kr/api/search/products?orderBy='+str(std)+'&category='+str(ctr)+''
        html=get(url,headers=headers).text
        src_tmp=json.loads(html)
        print(str(ctr)+str(std))
        for k in range(len(src_tmp)):
            pid.append(src_tmp[k]["$id"])
            pname.append(src_tmp[k]["name"])
            brandname.append(src_tmp[k]["brand"]["name"])
            p_score.append(str(src_tmp[k]["score"])[:3])
            num_of_reviews.append(src_tmp[k]["count"]["review"])
            category.append(pd_category[ctr])
```

```python
data ={'pid':pid,
      'pname':pname,
      'brandname':brandname,
      'num_of_reviews':num_of_reviews,
      'p_score':p_score,
      'category':category}
pd_id=DataFrame(data)
pd_id=DataFrame(data, columns=['category','brandname',"pid",'pname','p_score','num_of_reviews'])
pd_id
```

```python
pd_id=pd_id.drop_duplicates('pid')
```
