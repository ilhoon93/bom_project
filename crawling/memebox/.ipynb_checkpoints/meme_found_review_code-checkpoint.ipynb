```python
def starnum(star): # ë³„ì˜ í…ìŠ¤íŠ¸ë¥¼ ë½‘ì•„ë‚´ëŠ” í•¨ìˆ˜
    first=0
    second=0
    for i in range(len(star)):
        if star[i]=="\"":
            if first==0:
                first=i
            else: 
                second=i
    return star[first+1:second]
```

```python
score_dic ={"":0,"half":0.5,"one":1,"oneHalf":1.5,"two":2,"twoHalf":2.5,"three":3,"threeHalf":3.5,"four":4,"fourHalf":4.5,"five":5} 
    # ë³„ì ì´ ìˆ«ìê°€ ì•„ë‹Œ í…ìŠ¤íŠ¸ë¡œ ë˜ì–´ìˆìŒ. í…ìŠ¤íŠ¸ì— ëŒ€ì‘ë˜ëŠ” ë³„ì  ë”•ì…”ë„ˆë¦¬
```

```python
# ê¸°ë³¸ í™”ë©´ì—ì„œ reviewë°ì´í„° ë“¤ì–´ì˜¤ì§€ ì•ŠìŒ
# F12 ë„¤íŠ¸ì›Œí¬ ì°¾ì•„ì„œ ì‹¤ì‹œê°„ ì—…ë¡œë“œ ë˜ëŠ” íŒŒì¼ ê²€ìƒ‰
# ë¦¬ë·° ê´€ë ¨ íŒŒì¼ ì°¾ì•„ì„œ URL ì ‘ì†
# http://www.memebox.com/ajax/reviews?productId=511587&page=2&contentViewPickOption=2&hasBest=true
# page=x ì‹ìœ¼ë¡œ ë„£ì–´ì„œ í˜ì´ì§€ ì°¾ìŒ
item_id=[]
review_text=[] # ë¦¬ë·° í…ìŠ¤íŠ¸
created_date=[] # ë¦¬ë·° ì‘ì„±ì¼
star_score=[] # ë³„ì  
like_thumb_count=[] # ë¦¬ë·° ì¢‹ì•„ìš” ê°œìˆ˜
skin_type=[] # í”¼ë¶€íƒ€ì…
repurchase=[] # ì¬êµ¬ë§¤ì—¬ë¶€
```

```python
#### test
for i in meme_bb_id['item_id']:
    turl='http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page=1&contentViewPickOption=2&hasBest=true'
    thtml = get(turl,headers=headers).text
    tsoup=bs(thtml,'html.parser')
    if 50 < len(str(tsoup.select('div > a.arrow.last'))):
        ReviewPageNum=tsoup.select('div > a.arrow.last')[0].get('onclick')[15:17]
        ReviewPageNum=re.sub("'","",ReviewPageNum)
    else:
        ReviewPageNum=1
        
        # ë¦¬ë·° í˜ì´ì§€ ì´ ëª‡ ê°œ ì¸ì§€
        # í…ìŠ¤íŠ¸ ê¸¸ì´ë¡œ ë¦¬ë·° í˜ì´ì§€ 1 ì¸ì§€ ê·¸ ì´ìƒì¸ì§€ êµ¬ë¶„
        # "9'" ì˜ëª» ë“¤ì–´ì˜¨ í™‘ë”°ì˜´í‘œ ì œê±°
        # ë¦¬ë·° í˜ì´ì§€ ìˆ˜ë¥¼ ì–»ê¸° ìœ„í•œ ì„ì‹œ url,html, soup
        
    for j in range(1,int(ReviewPageNum)+1): # itemë³„ ë¦¬ë·° í˜ì´ì§€ì˜ ìˆ˜ ë§Œí¼ ë°˜ë³µ
        url = 'http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page='+str(j)+'&contentViewPickOption=2&hasBest=true'
        html = get(url,headers=headers).text
        soup=bs(html,'html.parser')
        
        for k in range(len(soup.select('div > p.reviews-written-text-deal-detail'))): # ë¦¬ë·° í˜ì´ì§€ë³„ ë¦¬ë·° ìˆ˜ ë§Œí¼ ë°˜ë³µ
            print("item_id " +soup.select('div.view')[k].get('data-product-id')) # item_id
            print("review_text "+re.sub('\t','',soup.select('div > p.reviews-written-text-deal-detail')[k].text).strip()) # ë¦¬ë·° í…ìŠ¤íŠ¸
            for i,child in enumerate(soup.select('div.info > span')[k]):
                if len(child)==4:
                    skin_type.append(re.sub("/","",child).strip())
                else:
                    skin_type.append("")
                         
            print("created_date "+ soup.select('div.info > span.date')[k].text) # ë¦¬ë·° ì‘ì„±ì¼ 
            print("star_score %d" % (score_dic[starnum(str(soup.select('span.star > em')[k]))])) # ë¦¬ë·° í‰ì 
            print("like_thumb_count "+re.sub("\n| ","",soup.select('span.like-thumb-text')[k].text)) # ë¦¬ë·° ì¢‹ì•„ìš” ê°œìˆ˜ 
            
```

```python
# run crawling
for i in meme_found__id['item_id']:
    turl='http://www.memebox.com/ajax/reviews?productId='+str(i)+'&contentViewPickOption=2&hasBest=true'
    thtml = get(turl,headers=headers).text
    tsoup=bs(thtml,'html.parser')
    if 50 < len(str(tsoup.select('div > a.arrow.last'))):
        ReviewPageNum=tsoup.select('div > a.arrow.last')[0].get('onclick')[15:17]
        ReviewPageNum=re.sub("'","",ReviewPageNum)
    else:
        ReviewPageNum=1
        
        # ë¦¬ë·° í˜ì´ì§€ ì´ ëª‡ ê°œ ì¸ì§€
        # í…ìŠ¤íŠ¸ ê¸¸ì´ë¡œ ë¦¬ë·° í˜ì´ì§€ 1 ì¸ì§€ ê·¸ ì´ìƒì¸ì§€ êµ¬ë¶„
        # "9'" ì˜ëª» ë“¤ì–´ì˜¨ í™‘ë”°ì˜´í‘œ ì œê±°
        # ë¦¬ë·° í˜ì´ì§€ ìˆ˜ë¥¼ ì–»ê¸° ìœ„í•œ ì„ì‹œ url,html, soup
        
    for j in range(1,int(ReviewPageNum)+1): # itemë³„ ë¦¬ë·° í˜ì´ì§€ì˜ ìˆ˜ ë§Œí¼ ë°˜ë³µ
        url = 'http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page='+str(j)+'&contentViewPickOption=2&hasBest=true'
        html = get(url,headers=headers).text
        soup=bs(html,'html.parser')
        
        for k in range(len(soup.select('div > p.reviews-written-text-deal-detail'))): # ë¦¬ë·° í˜ì´ì§€ë³„ ë¦¬ë·° ìˆ˜ ë§Œí¼ ë°˜ë³µ
            item_id.append(soup.select('div.view')[k].get('data-product-id')) # item_id
            review_text.append(re.sub('\t','',soup.select('div > p.reviews-written-text-deal-detail')[k].text).strip()) # ë¦¬ë·° í…ìŠ¤íŠ¸
            
            if len(soup.select('div.info')[k].select('span'))==2:
                skin_type.append(re.sub("/","",soup.select('div.info')[k].select('span')[0].text).strip()) #  í”¼ë¶€íƒ€ì… 
                created_date.append(soup.select('div.info')[k].select('span')[1].text)
            else:
                skin_type.append("") # í”¼ë¶€íƒ€ì… ì‘ì„± ì•ˆí–ˆìœ¼ë©´ ê³µë°±
                created_date.append(soup.select('div.info')[k].select('span')[0].text) # ë¦¬ë·° ì‘ì„±ì¼ 
     
            star_score.append(score_dic[starnum(str(soup.select('span.star > em')[k]))]) # ë¦¬ë·° í‰ì 
            like_thumb_count.append(re.sub("\n| ","",soup.select('span.like-thumb-text')[k].text)) # ë¦¬ë·° ì¢‹ì•„ìš” ê°œìˆ˜ 
                # ì •ê·œ í‘œí˜„ì‹ "\n| "ìœ¼ë¡œ \n í˜¹ì€ ê³µë°±ì„ ì œê±°
            if soup.select('div.review-selected-option')[k].select('span'):
                repurchase.append(soup.select('div.review-selected-option')[k].select('span')[0].text) # ì¬êµ¬ë§¤ ì—¬ë¶€
            else:
                repurchase.append("")
```

```python
data ={'item_id':item_id,
      'review_text':review_text,
      'created_date':created_date,
      'star_score':star_score,
      'like_thumb_count':like_thumb_count,
       'skin_type':skin_type,
       'repurchase':repurchase
    }
meme_found_review=DataFrame(data)
meme_found_review=DataFrame(data, columns=["item_id",'review_text','created_date','star_score','like_thumb_count','skin_type','repurchase'])
meme_found_review
```
