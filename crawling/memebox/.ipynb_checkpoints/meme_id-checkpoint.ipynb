```python
# run
def meme_id():
    dt=datetime.now() # 현재 날짜 dt에 저장
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
    
    urls={'http://www.memebox.com/category/345/review':"파운데이션",
         'http://www.memebox.com/category/343/review':"bb크림",
         'http://www.memebox.com/category/733/review':"cc크림"}
    for url in urls:
        driver.get(url)
        i=0
        while i<15:
            i = i+1
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(1)
        html = driver.page_source
        soup = bs(html, 'lxml')

        for i in range(len(soup.select('#container > div.newDealContents > ul.dealList'))):
            for item in soup.select("#container > div.newDealContents > ul.dealList")[i].find_all('li'):
                if str(dt.month) + "월" not in item.get("data-product-name") and "브랜드" not in item.get("data-product-name"):
                    if item.select('span > a > div.dealBottom > div.dealhead > div.starBox')==[]:
                        num_of_reviews.append(" ")
                    else:
                        num_of_reviews.append(item.select("span > a > div.dealBottom > div.dealhead > div.starBox > span.amount")[0].text[1:-1])

                    p_id.append(item.get("data-product-id"))
                    p_name.append(item.get("data-product-name"))
                    brandname.append(item.get("data-product-brand"))
                    p_score.append(mm_score_dic[mm_starnum(str(item.select("span > a > div.dealBottom > div.dealhead > div.starBox > span > em")))])
                    category.append(urls[url])
```
