```python
item_id=[] # 제품 id
item_name=[] # 제품 이름
review_text=[] # 리뷰 텍스트
created_date=[] # 리뷰 작성일
star_score=[] # 별점 
like_thumb_count=[] # 리뷰 좋아요 개수
skin_type=[] # 피부타입
repurchase=[] # 재구매여부
```

```python
## meme_new_review 

def meme_review_crawl():
    # 새로운 id 긁는 것은 프로토타입에서 안함 
    for p_id in meme_id['p_id']:
        url = 'https://www.memebox.com/ajax/reviews?productId='+str(p_id)+'&orderType=createdatDESC&page=1&contentViewPickOption=all'
        headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
        html = get(url,headers).text
        soup = bs(html,'lxml')
        
        t_date=datetime.today().strftime('%Y-%m-%d') # 오늘 
        yesterday = date.today() - timedelta(1)
        y_date=yesterday.strftime('%Y-%m-%d') # 어제
        
        
        for k in range(len(soup.select('body > form > div.view'))):
            r_date=date_transfer(soup.select('div.info')[k].select('span.date')[0].text)
            if r_date==t_date or r_date==y_date: # 리뷰 날짜가 오늘이나 어제와 같다면
                # item_name
                item_name.append(meme_id.loc[meme_id.p_id==p_id,'p_name'].values[0])

                # item_id
                item_id.append(meme_id.loc[meme_id.p_id==p_id,'p_id'].values[0]) # item_id

                #review_txt
                review_text.append(re.sub('\t','',soup.select('div > p.reviews-written-text-deal-detail')[k].text).strip()) # 리뷰 텍스트

                #  피부타입 & 리뷰 작성일
                if len(soup.select('div.info')[k].select('span'))==2:
                    skin_type.append(re.sub("/","",soup.select('div.info')[k].select('span')[0].text).strip())  
                    created_date.append(r_date)

                else:
                    skin_type.append(" ") # 피부타입 작성 안했으면 공백
                    created_date.append(r_date)


                # 리뷰 평점
                star_score.append(mm_score_dic[mm_starnum(str(soup.select('span.star > em')[k]))]) 

                # 리뷰 좋아요 개수
                like_thumb_count.append(re.sub("\n| ","",soup.select('span.like-thumb-text')[k].text))  
                    # 정규 표현식 "\n| "으로 \n 혹은 공백을 제거

                # 재구매 여부
                if soup.select('div.review-selected-option')[k].select('span'):
                    repurchase.append(soup.select('div.review-selected-option')[k].select('span')[0].text) # 재구매 여부
                else:
                    repurchase.append(" ")

```

```python
# auto 크롤러 실행, meme_new_review 를 만드는 리스트들 생성
meme_review_crawl() 
```

```python
data ={'pid':item_id,
      'text':review_text,
      'date':created_date,
      'score':star_score,
      'like_c':like_thumb_count,
       'skin_type':skin_type,
       'repurchased':repurchase,
       'pname':item_name
    }
meme_new_review=DataFrame(data)
meme_new_review=DataFrame(data, columns=["pid","pname",'text','date','score','like_c','skin_type','repurchased'])
meme_new_review # 새로 들어온 데이터를 new_review에 저장
```
