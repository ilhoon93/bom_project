```python
import pandas as pd
from pandas import Series, DataFrame
import re
import numpy as np
from requests import get
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import datetime
```

```python
meme_total_id=pd.read_csv('meme_total_id.csv')
```

```python
del meme_total_id['Unnamed: 0']
```

```python
meme_total_id #기존 저장된 미미박스 제품 id 
```

```python
# 별의 텍스트를 뽑아내는 함수
def starnum(star):
    first=0
    second=0
    for i in range(len(star)):
        if star[i]=="\"":
            if first==0:
                first=i
            else: 
                second=i
    return star[first+1:second]
```

```python
# 별점 텍스트에 대응되는 점수 딕셔너리로 저장
score_dic ={"":0,"one":1,"oneHalf":1.5,"two":2,"twoHalf":2.5,"three":3,"threeHalf":3.5,"four":4,"fourHalf":4.5,"five":5}

# 뽑아낸 데이터 저장할 리스트
item_id=[]
item_name=[]
item_brand=[]
item_score=[]
num_of_reviews=[]
category=[]
```

```python
## 미미박스 id 크롤링 

def meme_id_crawl():
    dt=datetime.datetime.now() # 현재 날짜 dt에 저장

    ## bb 크림 제품 id 크롤링
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
        # 사람인 것으로 속이는 header입력
        
    urls = {"http://www.memebox.com/category/343/newest":"bb크림",     # bb크림 
           "http://www.memebox.com/category/733/newest":"cc크림",      # cc크림
           "http://www.memebox.com/category/345/newest":"파운데이션"}  # 파운데이션
    for url in urls:
        html = get(url, headers=headers).text
        soup = bs(html, 'html.parser')
    
        for i in range(len(soup.select("#container > div.newDealContents > ul.dealList"))):
            for item in soup.select("#container > div.newDealContents > ul.dealList")[i].find_all('li'):
                if str(dt.month) + "월" not in item.get("data-product-name"):
                    if int(item.get("data-product-id")) not in list(meme_total_id['item_id']):
                        try:
                            
                            num_of_reviews.append(item.select("span > a > div.dealBottom > div.dealhead > div.starBox > span.amount")[0].text[1:-1])
                            item_id.append(item.get("data-product-id"))
                            item_name.append(item.get("data-product-name"))
                            item_brand.append(item.get("data-product-brand"))
                            item_score.append(score_dic[starnum(str(item.select("span > a > div.dealBottom > div.dealhead > div.starBox > span > em")))]) # 별점 평균
                            category.append(urls[url])
                        except:
                            continue
  
```

```python
meme_id_crawl()
```

```python
data ={'item_id':item_id,
      'item_name':item_name,
      'item_brand':item_brand,
      'num_of_reviews':num_of_reviews,
      'item_score':item_score,
      'category':category}
meme_new_id=DataFrame(data)
meme_new_id=DataFrame(data, columns=['category',"item_id",'item_brand','item_name','num_of_reviews','item_score'])
meme_new_id
```

```python
meme_new_id=meme_new_id.drop_duplicates(["item_id"]) # 같은 id 값을 제거
```
