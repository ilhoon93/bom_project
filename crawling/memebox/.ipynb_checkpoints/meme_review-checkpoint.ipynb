```python
# 기본 화면에서 review데이터 들어오지 않음
# F12 네트워크 찾아서 실시간 업로드 되는 파일 검색
# 리뷰 관련 파일 찾아서 URL 접속
# http://www.memebox.com/ajax/reviews?productId=511587&page=2&contentViewPickOption=2&hasBest=true
# page=x 식으로 넣어서 페이지 찾음
item_id=[]
item_name=[]
review_text=[] # 리뷰 텍스트
created_date=[] # 리뷰 작성일
star_score=[] # 별점 
like_thumb_count=[] # 리뷰 좋아요 개수
skin_type=[] # 피부타입
repurchase=[] # 재구매여부
```

```python
url='https://www.memebox.com/ajax/reviews?productId=523211&orderType=createdatDESC&page=160&contentViewPickOption=all'
html = get(url,headers=headers).text
soup=bs(html,'html.parser')

print(soup.select('div.view')[0].get('data-product-id')) # item_id
# 같은 제품 id  url로 접근했지만 페이지가 달라지면서 긁어오는 id값이 달라진다.
```

```python
# 페이지 수를 재코드, 가져오는 url을 변경, id에 저장된 개수와 가져온 리뷰의 개수가 다름!
# 같은 제품인데 예전에 판매했던 id 현재 판매하는 id가 다르다.
# 벗 같은 리뷰 페이지에 저장되어있다. 
# product_id를 긁어오는 코드로 접근하면 판매시 제품 id를 긁어와버린다.
# 현재 판매하는 id로 접근해도 리뷰 다 가져올 수 있다.
# 리뷰 id는 meme_id에서 가져와서 접근한 아이디를 그대로 저장

# 새롭게 짠 run crawl
for i in meme_id['p_id']:
    turl='http://www.memebox.com/ajax/reviews?productId='+str(i)+'&orderType=createdatDESC&contentViewPickOption=all'
    thtml = get(turl,headers=headers).text
    tsoup=bs(thtml,'lxml')
    
    page_num=int(tsoup.select('p.viewSelectText > span.pText')[0].text[:-1])
    page_num=math.ceil(page_num/10)
    
    
    for j in range(1,int(page_num)+1): # item별 리뷰 페이지의 수 만큼 반복
        url = 'http://www.memebox.com/ajax/reviews?productId='+str(i)+'&page='+str(j)+'&contentViewPickOption=2&hasBest=true'
        html = get(url,headers=headers).text
        soup=bs(html,'html.parser')

        for k in range(len(soup.select('div > p.reviews-written-text-deal-detail'))): # 리뷰 페이지별 리뷰 수 만큼 반복
            item_id.append(i) # item_id
            item_name.append(meme_id[meme_id.p_id==i]['p_name'].values)
            review_text.append(re.sub('\t','',soup.select('div > p.reviews-written-text-deal-detail')[k].text).strip()) # 리뷰 텍스트

            if len(soup.select('div.info')[k].select('span'))==2:
                skin_type.append(re.sub("/","",soup.select('div.info')[k].select('span')[0].text).strip()) #  피부타입 
                created_date.append(soup.select('div.info')[k].select('span')[1].text)
            else:
                skin_type.append("") # 피부타입 작성 안했으면 공백
                created_date.append(date_transfer(soup.select('div.info')[k].select('span')[0].text)) # 리뷰 작성일 

            star_score.append(mm_score_dic[mm_starnum(str(soup.select('span.star > em')[k]))]) # 리뷰 평점
            like_thumb_count.append(re.sub("\n| ","",soup.select('span.like-thumb-text')[k].text)) # 리뷰 좋아요 개수 
                # 정규 표현식 "\n| "으로 \n 혹은 공백을 제거
            if soup.select('div.review-selected-option')[k].select('span'):
                repurchase.append(soup.select('div.review-selected-option')[k].select('span')[0].text) # 재구매 여부
            else:
                repurchase.append("")
    
    
```

```python
print(len(item_id))
print(len(item_name))
print(len(review_text))
print(len(created_date))
print(len(star_score))
print(len(skin_type))
print(len(repurchase))
print(len(like_thumb_count))
```

```python
data ={'pid':item_id,
       'pname':item_name,
      'text':review_text,
      'date':created_date,
      'score':star_score,
      'like_c':like_thumb_count,
       'skin_type':skin_type,
       'repurchased':repurchase
    }
meme_total_review=DataFrame(data)
meme_total_review=DataFrame(data, columns=["pid",'pname','text','date','score','like_c','skin_type','repurchased'])
meme_total_review
```
